reading in factbook data to plot physics enrollment versus time.  This file also contains enrollments for other science majors.

    Columns:
    0 - year
    1 - physics
    2 - physics & female
    3 - chemistry
    4 - chem & female
    5 - math BS
    6 - math BS & female
    7 - CS
    8 - CS & female
    9 - Bio BS
    10 - Bio BS & female
    11 - Biochem
    12 - Biochem & female

```python
import numpy as np
from matplotlib import pyplot as plt
# the following command is for notebooks only
%matplotlib inline 

dat=np.loadtxt('Factbook.csv',delimiter=',',skiprows=2,usecols=np.arange(13))
```

```python
dat[:,0]
```

```python
plt.figure(figsize=(12,6))
plt.plot(dat[:,0],dat[:,1]-dat[:,2],'ko-',label='Physics Male',markersize=14)
plt.plot(dat[:,0],dat[:,2],'gs-',label='Physics Female',markersize=14)
plt.plot(dat[:,0],dat[:,2]/dat[:,1]*100,'g--',label= "% Women")
plt.legend(numpoints=1,loc='upper left',fontsize=12)
plt.xlabel('$Year$',fontsize=18)
plt.ylabel('$Number \ of \ Physics \ Majors$',fontsize=18)
plt.xticks((np.arange(1997,2021,2)),fontsize=14)
plt.yticks((np.arange(0,70,20)),fontsize=12)
plt.title('Siena College Physics Enrollment',fontsize=20)
#plt.axhline(y=35)
plt.savefig('physics-enrollment-gender-F2021.pdf')
```
